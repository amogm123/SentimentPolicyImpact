# Sentiment Analysis Demo
This notebook demonstrates the full pipeline for analyzing sentiment in policy feedback.

```python
# Load and preprocess data
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt
from wordcloud import WordCloud

nltk.download('punkt')
nltk.download('stopwords')

with open('../data/sample_feedback.txt') as f:
    lines = f.readlines()

text = ' '.join(lines)
tokens = word_tokenize(text.lower())
filtered = [w for w in tokens if w.isalpha() and w not in stopwords.words('english')]

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(filtered))
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
